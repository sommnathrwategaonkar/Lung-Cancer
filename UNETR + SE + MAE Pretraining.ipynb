```
# ===============================
# Single-page: UNETR + SE + MAE Pretraining + Extended Metrics + Plots
# ===============================
import os
import random
import math
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, TensorDataset, random_split

# sklearn metrics / display
from sklearn.metrics import (confusion_matrix, ConfusionMatrixDisplay,
                             roc_auc_score, average_precision_score,
                             matthews_corrcoef, cohen_kappa_score,
                             roc_curve, precision_recall_curve)

# -----------------------
# Config / device
# -----------------------
SEED = 42
random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
torch.cuda.manual_seed_all(SEED)

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

# -----------------------
# 1) Upload images (Colab)
# -----------------------
# If using local run: comment out upload block and set folder manually.
try:
    from google.colab import files
    os.makedirs("PNG_uploads", exist_ok=True)
    uploaded = files.upload()  # choose multiple PNG files
    for filename in uploaded.keys():
        os.rename(filename, os.path.join("PNG_uploads", filename))
    print(f"âœ… Uploaded {len(uploaded)} PNG images.")
except Exception:
    # assume folder already has images
    if not os.path.exists("PNG_uploads"):
        os.makedirs("PNG_uploads", exist_ok=True)
    print("Running outside Colab or upload skipped. Ensure PNGs are inside ./PNG_uploads")

# -----------------------
# 2) Precompute patches
# -----------------------
PATCH_SIZE = (128, 128)
NUM_PATCHES_PER_IMG = 10
PATCH_SAVE_PATH = "patches_tensor.pt"

class PNGPatchDataset(Dataset):
    def __init__(self, img_folder="PNG_uploads", patch_size=PATCH_SIZE, num_patches=NUM_PATCHES_PER_IMG):
        self.img_paths = [os.path.join(img_folder, f) for f in os.listdir(img_folder) if f.lower().endswith(".png")]
        self.patch_size = patch_size
        self.num_patches = num_patches
        self.samples = []
        for path in self.img_paths:
            img = Image.open(path).convert("L")
            img = np.array(img, dtype=np.float32)
            mn, mx = float(img.min()), float(img.max())
            if mx - mn < 1e-8:
                continue
            img = (img - mn) / (mx - mn)
            for _ in range(self.num_patches):
                self.samples.append(self.random_patch(img))

    def random_patch(self, img):
        h, w = img.shape
        ph, pw = self.patch_size
        if h < ph or w < pw:
            pad_h = max(0, ph - h)
            pad_w = max(0, pw - w)
            img = np.pad(img, ((0, pad_h), (0, pad_w)), mode="reflect")
            h, w = img.shape
        y = random.randint(0, h - ph)
        x = random.randint(0, w - pw)
        return img[y:y+ph, x:x+pw]

    def __len__(self): return len(self.samples)
    def __getitem__(self, idx):
        patch = torch.tensor(self.samples[idx][None, :, :], dtype=torch.float32)  # [1,H,W]
        return patch

# Build patches dataset and save to disk if not exists
if not os.path.exists(PATCH_SAVE_PATH):
    ds_png = PNGPatchDataset("PNG_uploads")
    if len(ds_png) == 0:
        raise RuntimeError("No valid patches extracted. Check uploads in ./PNG_uploads")
    all_patches = torch.stack([ds_png[i] for i in tqdm(range(len(ds_png)), desc="Stack patches")])
    torch.save(all_patches, PATCH_SAVE_PATH)
    print(f"âœ… {len(all_patches)} patches saved to {PATCH_SAVE_PATH}")
else:
    print(f"Loading existing patches from {PATCH_SAVE_PATH}")

# -----------------------
# 3) Load patches & DataLoaders
# -----------------------
patches = torch.load(PATCH_SAVE_PATH)
dataset = TensorDataset(patches, patches)  # pseudo target = input (for finetune use same as mask)
val_split = 0.2
val_len = max(1, int(len(dataset) * val_split))
train_len = max(1, len(dataset) - val_len)
train_ds, val_ds = random_split(dataset, [train_len, val_len])

BATCH_TRAIN = 8
BATCH_VAL = 4
pin = (device == "cuda")
train_loader = DataLoader(train_ds, batch_size=BATCH_TRAIN, shuffle=True, pin_memory=pin)
val_loader = DataLoader(val_ds, batch_size=BATCH_VAL, shuffle=False, pin_memory=pin)
print(f"âœ… Training patches: {len(train_ds)}, Validation patches: {len(val_ds)}")

# -----------------------
# 4) Model: UNETR + SE (from your uploaded code)
# -----------------------
class SEBlock(nn.Module):
    def __init__(self, channels, reduction=8):
        super().__init__()
        self.pool = nn.AdaptiveAvgPool2d(1)
        self.fc = nn.Sequential(
            nn.Conv2d(channels, channels // reduction, 1),
            nn.ReLU(inplace=True),
            nn.Conv2d(channels // reduction, channels, 1),
            nn.Sigmoid()
        )
    def forward(self, x):
        w = self.fc(self.pool(x))
        return x * w

class PatchEmbed(nn.Module):
    """Conv patchify: BxCxHxW -> BxNxD with non-overlapping patches."""
    def __init__(self, in_ch=1, embed_dim=256, patch_size=16):
        super().__init__()
        self.patch_size = patch_size
        self.proj = nn.Conv2d(in_ch, embed_dim, kernel_size=patch_size, stride=patch_size)
    def forward(self, x):
        y = self.proj(x)                         # B, D, H', W'
        B, D, Hp, Wp = y.shape
        tokens = y.flatten(2).transpose(1, 2)    # B, N, D
        return tokens, (Hp, Wp)

class TransformerBlock(nn.Module):
    def __init__(self, dim, num_heads=8, mlp_ratio=4.0, drop=0.0):
        super().__init__()
        self.norm1 = nn.LayerNorm(dim)
        self.attn  = nn.MultiheadAttention(dim, num_heads, dropout=drop, batch_first=True)
        self.norm2 = nn.LayerNorm(dim)
        self.mlp   = nn.Sequential(
            nn.Linear(dim, int(dim * mlp_ratio)),
            nn.GELU(),
            nn.Linear(int(dim * mlp_ratio), dim),
        )
    def forward(self, x):
        x = x + self.attn(self.norm1(x), self.norm1(x), self.norm1(x), need_weights=False)[0]
        x = x + self.mlp(self.norm2(x))
        return x

class ViTEncoder(nn.Module):
    def __init__(self, in_ch=1, embed_dim=256, depth=6, num_heads=8, patch_size=16):
        super().__init__()
        self.patch = PatchEmbed(in_ch, embed_dim, patch_size)
        self.pos = None
        self.blocks = nn.ModuleList([TransformerBlock(embed_dim, num_heads) for _ in range(depth)])
        self.embed_dim = embed_dim
        self.depth = depth
    def forward(self, x):
        tokens, grid = self.patch(x)       # B,N,D
        B, N, D = tokens.shape
        if (self.pos is None) or (self.pos.shape[1] != N):
            # create pos on device
            self.pos = nn.Parameter(torch.zeros(1, N, D, device=tokens.device))
            nn.init.trunc_normal_(self.pos, std=0.02)
        feats = []
        h = tokens + self.pos
        for i, blk in enumerate(self.blocks):
            h = blk(h)
            if i in {math.floor(self.depth*0.25)-1, math.floor(self.depth*0.5)-1,
                     math.floor(self.depth*0.75)-1, self.depth-1}:
                feats.append(h)
        return feats, grid  # list of B,N,D (all same token grid)

def tokens_to_map(tokens, grid_hw):
    B, N, D = tokens.shape
    H, W = grid_hw
    return tokens.transpose(1, 2).reshape(B, D, H, W)  # B,D,H',W'

class UNETR2D_SE(nn.Module):
    """
    UNETR-style decoder with SE at each stage.
    """
    def __init__(self, in_ch=1, out_ch=1, embed_dim=256, depth=6, patch_size=16):
        super().__init__()
        self.encoder = ViTEncoder(in_ch, embed_dim, depth=depth, num_heads=8, patch_size=patch_size)

        self.proj3 = nn.Conv2d(embed_dim, 256, 1)
        self.proj2 = nn.Conv2d(embed_dim, 128, 1)
        self.proj1 = nn.Conv2d(embed_dim, 64,  1)
        self.proj0 = nn.Conv2d(embed_dim, 32,  1)

        self.up3 = nn.ConvTranspose2d(256, 128, 2, stride=2)   # 8->16
        self.dec3 = nn.Sequential(nn.Conv2d(128,128,3,padding=1), nn.GELU(), SEBlock(128))

        self.up2 = nn.ConvTranspose2d(128, 64, 2, stride=2)    # 16->32
        self.dec2 = nn.Sequential(nn.Conv2d(64,64,3,padding=1), nn.GELU(), SEBlock(64))

        self.up1 = nn.ConvTranspose2d(64, 32, 2, stride=2)     # 32->64
        self.dec1 = nn.Sequential(nn.Conv2d(32,32,3,padding=1), nn.GELU(), SEBlock(32))

        self.final_up = nn.ConvTranspose2d(32, 32, 2, stride=2) # 64->128
        self.head = nn.Conv2d(32, out_ch, 1)

        self.patch_size = patch_size

    def forward(self, x):
        feats, grid = self.encoder(x)           # f0,f1,f2,f3 (token seqs)
        f0, f1, f2, f3 = [tokens_to_map(f, grid) for f in feats]  # all at 8x8 if patch=16 for 128x128

        # Stage @8x8 -> @16x16
        x = self.proj3(f3)                      # 8x8
        x = self.up3(x)                         # 16x16
        skip2 = F.interpolate(self.proj2(f2), size=x.shape[-2:], mode="bilinear", align_corners=False)
        x = self.dec3(x + skip2)

        # @16x16 -> @32x32
        x = self.up2(x)                         # 32x32
        skip1 = F.interpolate(self.proj1(f1), size=x.shape[-2:], mode="bilinear", align_corners=False)
        x = self.dec2(x + skip1)

        # @32x32 -> @64x64
        x = self.up1(x)                         # 64x64
        skip0 = F.interpolate(self.proj0(f0), size=x.shape[-2:], mode="bilinear", align_corners=False)
        x = self.dec1(x + skip0)

        # @64x64 -> @128x128
        x = self.final_up(x)                    # 128x128
        logits = self.head(x)                   # logits (no sigmoid)
        return logits

# -----------------------
# 5) MAE-style pretraining utilities (encoder-only MAE)
# -----------------------
class MAEHead(nn.Module):
    def __init__(self, embed_dim=256, patch_size=16):
        super().__init__()
        self.patch_size = patch_size
        self.proj = nn.Linear(embed_dim, patch_size * patch_size)  # grayscale
    def forward(self, tokens):  # B,N,D -> B,N,P*P
        return self.proj(tokens)

def mae_random_mask(x, mask_ratio=0.6):
    B, N, D = x.shape
    len_keep = int(N * (1 - mask_ratio))
    noise = torch.rand(B, N, device=x.device)
    ids_shuffle = torch.argsort(noise, dim=1)
    ids_restore = torch.argsort(ids_shuffle, dim=1)
    ids_keep = ids_shuffle[:, :len_keep]
    x_keep = torch.gather(x, 1, ids_keep.unsqueeze(-1).expand(-1, -1, D))
    mask = torch.ones(B, N, device=x.device)
    mask[:, :len_keep] = 0
    mask = torch.gather(mask, 1, ids_restore)
    return x_keep, mask, ids_restore

def mae_unshuffle(tokens_keep, ids_restore):
    B, Nk, D = tokens_keep.shape
    N = ids_restore.shape[1]
    # Ensure same dtype as tokens_keep to avoid scatter dtype mismatch
    tokens_full = torch.zeros(B, N, D, device=tokens_keep.device, dtype=tokens_keep.dtype)
    tokens_full.scatter_(1, ids_restore[:, :Nk].unsqueeze(-1).expand(-1, -1, D), tokens_keep)
    return tokens_full

@torch.no_grad()
def _ensure_pos(encoder, N, device_):
    if (encoder.pos is None) or (encoder.pos.shape[1] != N):
        encoder.pos = nn.Parameter(torch.zeros(1, N, encoder.embed_dim, device=device_))
        nn.init.trunc_normal_(encoder.pos, std=0.02)

def pretrain_mae(unetr_model, dataloader, val_loader=None, epochs=5, mask_ratio=0.6, lr=1e-4):
    encoder = unetr_model.encoder
    mae_head = MAEHead(embed_dim=encoder.embed_dim, patch_size=unetr_model.patch_size).to(device)
    opt = torch.optim.AdamW(list(encoder.parameters()) + list(mae_head.parameters()), lr=lr, weight_decay=1e-4)
    scaler = torch.amp.GradScaler(enabled=(device == "cuda"))

    train_losses, val_losses = [], []

    encoder.train()
    for ep in range(epochs):
        total = 0.0
        for (imgs, _) in tqdm(dataloader, desc=f"MAE Pretrain {ep+1}/{epochs}"):
            imgs = imgs.to(device, non_blocking=True)
            opt.zero_grad(set_to_none=True)
            with torch.amp.autocast(device_type="cuda", enabled=(device == "cuda")):
                tokens, grid = encoder.patch(imgs)             # B,N,D
                B, N, D = tokens.shape
                _ensure_pos(encoder, N, tokens.device)
                h = tokens + encoder.pos
                for blk in encoder.blocks:
                    h = blk(h)

                tokens_keep, mask, ids_restore = mae_random_mask(h, mask_ratio=mask_ratio)
                pred_keep = mae_head(tokens_keep)              # B, Nk, P*P
                pred_full = mae_unshuffle(pred_keep, ids_restore)  # B, N, P*P

                P = unetr_model.patch_size
                gt = F.unfold(imgs, kernel_size=P, stride=P)   # B, 1*P*P, N
                gt = gt.transpose(1, 2).squeeze(2)             # B, N, P*P

                mask_expand = mask.unsqueeze(-1)               # B, N, 1
                loss = ((pred_full - gt) ** 2 * mask_expand).sum() / (mask_expand.sum() * pred_full.shape[-1] + 1e-8)

            scaler.scale(loss).backward()
            scaler.step(opt)
            scaler.update()
            total += loss.item()

        avg_train = total / max(len(dataloader), 1)
        train_losses.append(avg_train)

        # validation during MAE pretraining (optional)
        if val_loader is not None:
            encoder.eval()
            val_total = 0.0
            with torch.no_grad():
                for (imgs, _) in val_loader:
                    imgs = imgs.to(device, non_blocking=True)
                    with torch.amp.autocast(device_type="cuda", enabled=(device == "cuda")):
                        tokens, grid = encoder.patch(imgs)
                        B, N, D = tokens.shape
                        _ensure_pos(encoder, N, tokens.device)
                        h = tokens + encoder.pos
                        for blk in encoder.blocks:
                            h = blk(h)
                        tokens_keep, mask, ids_restore = mae_random_mask(h, mask_ratio=mask_ratio)
                        pred_keep = mae_head(tokens_keep)
                        pred_full = mae_unshuffle(pred_keep, ids_restore)
                        P = unetr_model.patch_size
                        gt = F.unfold(imgs, kernel_size=P, stride=P)
                        gt = gt.transpose(1, 2).squeeze(2)
                        mask_expand = mask.unsqueeze(-1)
                        loss = ((pred_full - gt) ** 2 * mask_expand).sum() / (mask_expand.sum() * pred_full.shape[-1] + 1e-8)
                    val_total += loss.item()
            avg_val = val_total / max(len(val_loader), 1)
            val_losses.append(avg_val)
            print(f"MAE Epoch {ep+1}/{epochs} | Train Loss: {avg_train:.6f} | Val Loss: {avg_val:.6f}")
        else:
            print(f"MAE Epoch {ep+1}/{epochs} | Train Loss: {avg_train:.6f}")

        # save checkpoint of encoder+mae_head at each epoch
        torch.save({
            "encoder_state": encoder.state_dict(),
            "mae_head_state": mae_head.state_dict(),
            "epoch": ep+1
        }, f"mae_checkpoint_epoch{ep+1}.pth")

    # plot losses
    plt.figure(figsize=(7,5)); plt.plot(train_losses, label="mae_train", marker='o')
    if len(val_losses)>0: plt.plot(val_losses, label="mae_val", marker='s')
    plt.xlabel("Epoch"); plt.ylabel("MSE Loss"); plt.title("MAE Pretraining Loss"); plt.legend(); plt.grid(True); plt.show()

    return

# -----------------------
# 6) Metrics helpers for finetune
# -----------------------
def _sigmoid(x): return torch.sigmoid(x)

def dice_coef(logits, target, eps=1e-6):
    pred = (_sigmoid(logits) > 0.5).float()
    target = (target > 0.5).float()
    num = (2 * (pred * target).sum(dim=(1,2,3)))
    den = (pred.sum(dim=(1,2,3)) + target.sum(dim=(1,2,3)) + eps)
    return (num / den).mean().item()

def iou_score(logits, target, eps=1e-6):
    pred = (_sigmoid(logits) > 0.5).float()
    target = (target > 0.5).float()
    inter = (pred * target).sum(dim=(1,2,3))
    union = (pred + target).sum(dim=(1,2,3)) - inter
    return (inter / (union + eps)).mean().item()

def accuracy(logits, target):
    pred = (_sigmoid(logits) > 0.5).float()
    return (pred == (target > 0.5).float()).float().mean().item()

def precision(logits, target, eps=1e-6):
    pred = (_sigmoid(logits) > 0.5).float()
    target = (target > 0.5).float()
    tp = (pred * target).sum()
    fp = (pred * (1 - target)).sum()
    return (tp / (tp + fp + eps)).item()

def recall(logits, target, eps=1e-6):
    pred = (_sigmoid(logits) > 0.5).float()
    target = (target > 0.5).float()
    tp = (pred * target).sum()
    fn = (((1 - pred) * target)).sum()
    return (tp / (tp + fn + eps)).item()

def f1_score(logits, target, eps=1e-6):
    p = precision(logits,target,eps)
    r = recall(logits,target,eps)
    return 2*p*r/(p+r+eps)

def confusion_matrix_stats(logits, target, threshold=0.5):
    pred = (_sigmoid(logits) > threshold).cpu().numpy().astype(int).ravel()
    target = (target > 0.5).cpu().numpy().astype(int).ravel()
    cm = confusion_matrix(target, pred, labels=[0,1])
    tn, fp, fn, tp = cm.ravel()
    return int(tp), int(tn), int(fp), int(fn), cm

# -----------------------
# 7) Build model, pretrain (MAE), then fine-tune with extended metrics + plots
# -----------------------
model = UNETR2D_SE(in_ch=1, out_ch=1, embed_dim=256, depth=6, patch_size=16).to(device)

# 7.a Pretrain MAE on train_loader (encoder-level)
PRETRAIN_EPOCHS = 5     # increase for stronger pretraining
MASK_RATIO = 0.6
print("ðŸš€ Starting MAE pretraining...")
pretrain_mae(model, train_loader, val_loader=None, epochs=PRETRAIN_EPOCHS, mask_ratio=MASK_RATIO, lr=1e-4)
print("âœ… MAE pretraining done.")

# 7.b Fine-tune full model
criterion = nn.BCEWithLogitsLoss()  # use logits -> numerically stable
optimizer = torch.optim.Adam(model.parameters(), lr=1e-4, weight_decay=1e-4)
EPOCHS_FT = 20

best_val = float("inf")
for epoch in range(EPOCHS_FT):
    # ---- Train ----
    model.train()
    running = 0.0
    for imgs, masks in tqdm(train_loader, desc=f"FT Epoch {epoch+1}/{EPOCHS_FT} - Train"):
        imgs, masks = imgs.to(device, non_blocking=True), masks.to(device, non_blocking=True)
        optimizer.zero_grad(set_to_none=True)
        logits = model(imgs)
        loss = criterion(logits, masks)
        loss.backward()
        optimizer.step()
        running += loss.item()
    train_loss = running / max(len(train_loader), 1)

    # ---- Validate ----
    model.eval()
    val_loss = dice = iou = acc = prec = rec = f1 = 0.0
    tp_total = tn_total = fp_total = fn_total = 0
    all_probs, all_preds, all_targets = [], [], []
    with torch.no_grad():
        for imgs, masks in val_loader:
            imgs, masks = imgs.to(device, non_blocking=True), masks.to(device, non_blocking=True)
            logits = model(imgs)
            val_loss += criterion(logits, masks).item()
            dice += dice_coef(logits, masks)
            iou  += iou_score(logits, masks)
            acc  += accuracy(logits, masks)
            prec += precision(logits, masks)
            rec  += recall(logits, masks)
            f1   += f1_score(logits, masks)

            probs = torch.sigmoid(logits).cpu().numpy().ravel()
            preds = (probs > 0.5).astype(int)
            targs = (masks.cpu().numpy().ravel() > 0.5).astype(int)

            all_probs.extend(probs.tolist())
            all_preds.extend(preds.tolist())
            all_targets.extend(targs.tolist())

            tp, tn, fp, fn, cm = confusion_matrix_stats(logits, masks)
            tp_total += tp; tn_total += tn; fp_total += fp; fn_total += fn

    nv = max(len(val_loader), 1)
    val_loss /= nv; dice /= nv; iou /= nv; acc /= nv; prec /= nv; rec /= nv; f1 /= nv

    # Extra metrics (robust to single-class cases)
    try:
        roc_auc = roc_auc_score(all_targets, all_probs) if len(set(all_targets)) > 1 else float("nan")
    except Exception:
        roc_auc = float("nan")
    try:
        pr_auc = average_precision_score(all_targets, all_probs) if len(set(all_targets)) > 1 else float("nan")
    except Exception:
        pr_auc = float("nan")
    try:
        mcc = matthews_corrcoef(all_targets, all_preds) if len(set(all_preds)) > 1 and len(set(all_targets))>1 else float("nan")
    except Exception:
        mcc = float("nan")
    try:
        kappa = cohen_kappa_score(all_targets, all_preds) if len(set(all_preds)) > 1 and len(set(all_targets))>1 else float("nan")
    except Exception:
        kappa = float("nan")

    spec = tn_total / (tn_total + fp_total + 1e-8)
    npv_val = tn_total / (tn_total + fn_total + 1e-8)

    print(f"Epoch {epoch+1}/{EPOCHS_FT} -> "
          f"Train {train_loss:.6f} | Val {val_loss:.6f} | "
          f"Dice {dice:.4f} | IoU {iou:.4f} | Acc {acc:.4f} | "
          f"Prec {prec:.4f} | Rec {rec:.4f} | F1 {f1:.4f} | "
          f"Spec {spec:.4f} | NPV {npv_val:.4f} | "
          f"ROC-AUC {roc_auc:.4f} | PR-AUC {pr_auc:.4f} | "
          f"MCC {mcc:.4f} | Kappa {kappa:.4f}")

    # ---- Confusion matrix plot & save ----
    all_cm = confusion_matrix(all_targets, all_preds, labels=[0,1])
    disp = ConfusionMatrixDisplay(confusion_matrix=all_cm, display_labels=["Background","Nodule"])
    disp.plot(cmap="Blues", values_format="d")
    plt.title(f"Validation Confusion Matrix - Epoch {epoch+1}")
    plt.savefig(f"cm_epoch{epoch+1}.png", bbox_inches="tight", dpi=150)
    plt.show()

    # ---- ROC and PR curve plots if possible ----
    if len(set(all_targets)) > 1:
        fpr, tpr, _ = roc_curve(all_targets, all_probs)
        plt.figure(figsize=(6,5)); plt.plot(fpr, tpr); plt.xlabel("FPR"); plt.ylabel("TPR"); plt.title(f"ROC Curve - Epoch {epoch+1} (AUC={roc_auc:.3f})"); plt.grid(True)
        plt.savefig(f"roc_epoch{epoch+1}.png", bbox_inches="tight", dpi=150); plt.show()

        precs, recs, _ = precision_recall_curve(all_targets, all_probs)
        plt.figure(figsize=(6,5)); plt.plot(recs, precs); plt.xlabel("Recall"); plt.ylabel("Precision"); plt.title(f"PR Curve - Epoch {epoch+1} (AUC={pr_auc:.3f})"); plt.grid(True)
        plt.savefig(f"pr_epoch{epoch+1}.png", bbox_inches="tight", dpi=150); plt.show()

    # ---- Save best / last model ----
    if val_loss < best_val:
        best_val = val_loss
        torch.save(model.state_dict(), "unetr2d_se_mae_best.pt")
    torch.save(model.state_dict(), f"unetr2d_se_mae_epoch{epoch+1}.pt")
    torch.save(model.state_dict(), "unetr2d_se_mae_last.pt")

print("âœ… Saved: unetr2d_se_mae_best.pt and unetr2d_se_mae_last.pt")

```

```
import os, glob, random
import numpy as np
from PIL import Image
import torch, torch.nn as nn, torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from tqdm import tqdm
import matplotlib.pyplot as plt
from google.colab import files

# ================= CONFIG =================
class CFG:
    seed=42; img_size=256; in_ch,out_ch=1,1
    batch_size=2; epochs=30; lr=1e-3; val_split=0.2
    dropout_p=0.3; patience=5
    device='cuda' if torch.cuda.is_available() else 'cpu'
    out_dir='./outputs_swinunet_cbam_mcdo'
    os.makedirs(out_dir, exist_ok=True)

random.seed(CFG.seed); np.random.seed(CFG.seed)
torch.manual_seed(CFG.seed)
if torch.cuda.is_available(): torch.cuda.manual_seed_all(CFG.seed)

# ================= Upload + Auto-Mask =================
def manual_upload_auto_mask():
    uploaded=files.upload()
    os.makedirs('./data/images',exist_ok=True); os.makedirs('./data/masks',exist_ok=True)
    for fn,data in uploaded.items():
        path=f'./data/images/{fn}'; open(path,'wb').write(data)
        img=Image.open(path).convert('L').resize((CFG.img_size,CFG.img_size))
        arr=np.array(img)/255.; mask=(arr>0.5).astype(np.uint8)*255
        Image.fromarray(mask).save(f'./data/masks/{os.path.splitext(fn)[0]}_mask.png')
    print("Images uploaded; synthetic masks generated.")

# ================= Dataset =================
class SegDataset(Dataset):
    def __init__(self,pairs): self.pairs=pairs
    def __len__(self): return len(self.pairs)
    def __getitem__(self,idx):
        img_path,msk_path=self.pairs[idx]
        img=Image.open(img_path).convert('L').resize((CFG.img_size,CFG.img_size))
        msk=Image.open(msk_path).convert('L').resize((CFG.img_size,CFG.img_size))
        img,msk=np.array(img)/255., np.array(msk)/255.
        return torch.tensor(img).float().unsqueeze(0), torch.tensor(msk).float().unsqueeze(0)

def discover_and_pair():
    imgs=glob.glob('./data/images/*'); msks=glob.glob('./data/masks/*')
    img_dict={os.path.splitext(os.path.basename(f))[0]:f for f in imgs}
    msk_dict={os.path.splitext(os.path.basename(f))[0].replace('_mask',''):f for f in msks}
    pairs=[(img_dict[k],msk_dict[k]) for k in img_dict if k in msk_dict]
    print(f"Images: {len(imgs)}, Masks: {len(msks)}, Pairs: {len(pairs)}")
    return pairs

# ================= Model =================
class CBAM(nn.Module):
    def __init__(self,ch,r=16):
        super().__init__()
        self.mlp=nn.Sequential(nn.Linear(ch,ch//r),nn.ReLU(),nn.Linear(ch//r,ch))
        self.conv=nn.Conv2d(2,1,7,padding=3)
    def forward(self,x):
        b,c,h,w=x.size(); y,z=x.mean((2,3),True),x.amax((2,3),True)
        att=torch.sigmoid(self.mlp(y.view(b,c))+self.mlp(z.view(b,c))).view(b,c,1,1)
        x=x*att; avg,max_=torch.mean(x,1,True),torch.max(x,1,True)[0]
        att2=torch.sigmoid(self.conv(torch.cat([avg,max_],1))); return x*att2

class BasicBlock(nn.Module):
    def __init__(self,ch):
        super().__init__()
        self.conv=nn.Sequential(nn.Conv2d(ch,ch,3,1,1),nn.BatchNorm2d(ch),nn.ReLU(),
                                nn.Conv2d(ch,ch,3,1,1),nn.BatchNorm2d(ch),nn.ReLU(),
                                nn.Dropout(CFG.dropout_p))
        self.cbam=CBAM(ch)
    def forward(self,x): return self.cbam(self.conv(x))

class Down(nn.Module):
    def __init__(self,in_ch,out_ch):
        super().__init__(); self.mp=nn.MaxPool2d(2); self.conv=nn.Conv2d(in_ch,out_ch,3,1,1)
        self.block=BasicBlock(out_ch)
    def forward(self,x): return self.block(self.conv(self.mp(x)))

class Up(nn.Module):
    def __init__(self,in_ch,out_ch):
        super().__init__(); self.up=nn.ConvTranspose2d(in_ch,in_ch//2,2,2); self.conv=nn.Conv2d(in_ch,out_ch,3,1,1)
        self.block=BasicBlock(out_ch)
    def forward(self,x1,x2):
        x1=self.up(x1); diffY,diffX=x2.size()[2]-x1.size()[2],x2.size()[3]-x1.size()[3]
        x1=F.pad(x1,[diffX//2,diffX-diffX//2,diffY//2,diffY-diffY//2])
        return self.block(self.conv(torch.cat([x2,x1],1)))

class OutConv(nn.Module):
    def __init__(self,in_ch,out_ch): super().__init__(); self.conv=nn.Conv2d(in_ch,out_ch,1)
    def forward(self,x): return self.conv(x)

class SwinUNet_CBAM(nn.Module):
    def __init__(self):
        super().__init__()
        self.inconv=nn.Conv2d(CFG.in_ch,64,3,1,1); self.block1=BasicBlock(64)
        self.down1=Down(64,128); self.down2=Down(128,256); self.down3=Down(256,512)
        self.up1=Up(512,256); self.up2=Up(256,128); self.up3=Up(128,64)
        self.outc=OutConv(64,CFG.out_ch)
    def forward(self,x):
        x1=self.block1(self.inconv(x)); x2=self.down1(x1); x3=self.down2(x2); x4=self.down3(x3)
        x=self.up1(x4,x3); x=self.up2(x,x2); x=self.up3(x,x1)
        return torch.sigmoid(self.outc(x))

# ================= Metrics =================
def dice(pred,target,eps=1e-6): pred=(pred>0.5).float(); return (2*(pred*target).sum()+eps)/(pred.sum()+target.sum()+eps)
def iou(pred,target,eps=1e-6): pred=(pred>0.5).float(); inter=(pred*target).sum(); return (inter+eps)/(pred.sum()+target.sum()-inter+eps)

# ================= Train Loop =================
def train_model(model,tr_loader,va_loader):
    opt=torch.optim.Adam(model.parameters(),lr=CFG.lr)
    sched=torch.optim.lr_scheduler.CosineAnnealingWarmRestarts(opt,T_0=5)
    best_dice=0; patience=0; last_metrics={}
    for ep in range(CFG.epochs):
        model.train(); tloss=0
        for imgs,masks in tqdm(tr_loader,desc=f'FT Epoch {ep+1}/{CFG.epochs}',unit='batch'):
            imgs,masks=imgs.to(CFG.device),masks.to(CFG.device)
            opt.zero_grad(); out=model(imgs)
            loss=F.binary_cross_entropy(out,masks); loss.backward(); opt.step(); tloss+=loss.item()
        tloss/=len(tr_loader)

        # Validation
        model.eval(); all_preds=[]; all_tgts=[]; dices=[]; ious=[]
        with torch.no_grad():
            for imgs,masks in va_loader:
                imgs,masks=imgs.to(CFG.device),masks.to(CFG.device)
                out=model(imgs)
                dices.append(dice(out,masks).item()); ious.append(iou(out,masks).item())
                all_preds.extend(out.cpu().numpy().ravel()); all_tgts.extend(masks.cpu().numpy().ravel())
        all_tgts_bin=(np.array(all_tgts)>0.5).astype(int)
        all_preds_bin=np.round(all_preds).astype(int)
        acc=accuracy_score(all_tgts_bin,all_preds_bin)
        prec=precision_score(all_tgts_bin,all_preds_bin,zero_division=0)
        rec=recall_score(all_tgts_bin,all_preds_bin,zero_division=0)
        f1=f1_score(all_tgts_bin,all_preds_bin,zero_division=0)
        mdice,miou=np.mean(dices),np.mean(ious)
        print(f"Epoch {ep+1}/{CFG.epochs} -> Train {tloss:.4f} | Dice {mdice:.4f} | IoU {miou:.4f} | Acc {acc:.4f} | Prec {prec:.4f} | Rec {rec:.4f} | F1 {f1:.4f}")
        last_metrics={'acc':acc,'prec':prec,'rec':rec,'f1':f1,'cm':confusion_matrix(all_tgts_bin,all_preds_bin)}
        if mdice>best_dice: best_dice=mdice; patience=0; torch.save(model.state_dict(),f"{CFG.out_dir}/best.pth")
        else: patience+=1
        if patience>=CFG.patience: print("Early stopping"); break

    # ====== Plot CM ======
    cm=last_metrics['cm']; plt.figure(figsize=(4,4))
    plt.imshow(cm,interpolation='nearest',cmap=plt.cm.Blues); plt.title('Confusion Matrix')
    plt.colorbar(); tick_marks=np.arange(2)
    plt.xticks(tick_marks,['BG','Lesion']); plt.yticks(tick_marks,['BG','Lesion'])
    thresh=cm.max()/2
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j,i,f"{cm[i,j]}",ha='center',color='white' if cm[i,j]>thresh else 'black')
    plt.ylabel('True'); plt.xlabel('Pred'); plt.tight_layout(); plt.show()

    # ====== Plot metrics bar ======
    metrics=['Accuracy','Precision','Recall','F1']; values=[last_metrics['acc'],last_metrics['prec'],last_metrics['rec'],last_metrics['f1']]
    plt.figure(figsize=(6,4)); plt.bar(metrics,values,color=['skyblue','orange','green','red'])
    plt.ylim(0,1); plt.title('Last Epoch Metrics'); plt.ylabel('Score'); plt.show()
    print(f"Last Epoch Metrics -> Acc: {last_metrics['acc']:.4f} | Prec: {last_metrics['prec']:.4f} | Rec: {last_metrics['rec']:.4f} | F1: {last_metrics['f1']:.4f}")

# ================= Main =================
def main():
    manual_upload_auto_mask()
    pairs=discover_and_pair(); random.shuffle(pairs)
    split=int(len(pairs)*(1-CFG.val_split))
    tr_loader=DataLoader(SegDataset(pairs[:split]),batch_size=CFG.batch_size,shuffle=True)
    va_loader=DataLoader(SegDataset(pairs[split:]),batch_size=CFG.batch_size)
    model=SwinUNet_CBAM().to(CFG.device)
    train_model(model,tr_loader,va_loader)
    print("Done. Outputs in",CFG.out_dir)

if __name__=='__main__': main()
```

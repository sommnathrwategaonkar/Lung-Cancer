```
# =============================== #
# SwinUNet + CBAM + MC-Dropout    #
# =============================== #
import os, glob, random, io
import numpy as np
from PIL import Image
from tqdm import tqdm
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader

from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
)

# -------- Colab upload (safe to import elsewhere) --------
try:
    from google.colab import files
    IN_COLAB = True
except Exception:
    IN_COLAB = False

# ================= CONFIG =================
class CFG:
    seed        = 42
    img_size    = 256
    in_ch       = 1
    out_ch      = 1
    batch_size  = 4
    epochs      = 20
    lr          = 1e-3
    val_split   = 0.2
    dropout_p   = 0.3           # used by blocks -> enables MC-Dropout
    patience    = 5
    mc_passes   = 8             # T stochastic passes for MC-Dropout evaluation
    device      = 'cuda' if torch.cuda.is_available() else 'cpu'
    data_root   = './data'
    out_dir     = './outputs_swinunet_cbam_mcdo'
    os.makedirs(out_dir, exist_ok=True)
    # folders the script manages
    img_dir     = os.path.join(data_root, 'images')
    msk_dir     = os.path.join(data_root, 'masks')
    vimg_dir    = os.path.join(data_root, 'val_images')
    vmsk_dir    = os.path.join(data_root, 'val_masks')

# seeds
random.seed(CFG.seed); np.random.seed(CFG.seed)
torch.manual_seed(CFG.seed)
if torch.cuda.is_available(): torch.cuda.manual_seed_all(CFG.seed)
print(f"Device: {CFG.device}")

# ================= File helpers =================
def _ensure_dirs():
    for d in [CFG.img_dir, CFG.msk_dir, CFG.vimg_dir, CFG.vmsk_dir]:
        os.makedirs(d, exist_ok=True)

def _pil_to_arr_gray(pil_img, size=CFG.img_size):
    return np.array(pil_img.convert('L').resize((size, size)), dtype=np.uint8)

def _auto_mask_from_image(arr_uint8, thresh=128):
    # Basic synthetic mask via global threshold
    return (arr_uint8 > thresh).astype(np.uint8) * 255

def upload_images_and_optional_masks():
    """
    Colab-only helper: user selects arbitrary files.
    Images = .png/.jpg/.jpeg (case-insensitive)
    Masks   = file names that end with '_mask' OR placed manually in masks folder after this.
    If no masks provided, synthetic masks are generated into ./data/masks.
    """
    if not IN_COLAB:
        print("Not in Colab: skip interactive upload. Make sure data exists in ./data/... folders.")
        return

    _ensure_dirs()
    print("üì§ Select images (and optionally masks). You can multi-select.")
    uploaded = files.upload()
    if len(uploaded) == 0:
        print("‚ö†Ô∏è No files uploaded.")
        return

    img_count, msk_count = 0, 0
    # naive rule: treat files with '_mask' in name as masks; the rest as images
    for fn, raw in uploaded.items():
        ext = os.path.splitext(fn)[1].lower()
        if ext not in ['.png', '.jpg', '.jpeg']:
            print(f"Skipping non-image: {fn}")
            continue

        is_mask = ('_mask' in os.path.splitext(fn)[0].lower())
        target_dir = CFG.msk_dir if is_mask else CFG.img_dir

        with open(os.path.join(target_dir, os.path.basename(fn)), 'wb') as f:
            f.write(raw)
        if is_mask: msk_count += 1
        else: img_count += 1

    print(f"‚úÖ Uploaded {img_count} images and {msk_count} masks.")

    # If there are images with no masks, auto-generate simple masks
    img_basenames = {os.path.splitext(os.path.basename(p))[0] for p in glob.glob(os.path.join(CFG.img_dir, '*'))}
    msk_basenames = {os.path.splitext(os.path.basename(p))[0].replace('_mask','') for p in glob.glob(os.path.join(CFG.msk_dir, '*'))}

    need_masks = list(img_basenames - msk_basenames)
    if len(need_masks) > 0:
        print(f"üõ† Generating {len(need_masks)} synthetic masks (thresholding) ...")
        for base in need_masks:
            img_path = os.path.join(CFG.img_dir, base + '.png')
            if not os.path.exists(img_path):  # try jpg/jpeg
                for ext in ['.jpg', '.jpeg', '.JPG', '.JPEG', '.PNG', '.png']:
                    alt = os.path.join(CFG.img_dir, base + ext)
                    if os.path.exists(alt):
                        img_path = alt; break
            if not os.path.exists(img_path):
                print(f"  ‚ö†Ô∏è Could not find image file for {base}, skipping.")
                continue
            arr = _pil_to_arr_gray(Image.open(img_path))
            msk = _auto_mask_from_image(arr)
            Image.fromarray(msk).save(os.path.join(CFG.msk_dir, base + '_mask.png'))
        print("‚úÖ Synthetic masks generated.")

# ================= Pair discovery =================
def discover_pairs(image_dir, mask_dir):
    """
    Return list of (img_path, msk_path) where mask is matched either by:
    - same basename; or
    - basename + '_mask'
    """
    imgs = sorted([p for p in glob.glob(os.path.join(image_dir, '*')) if os.path.splitext(p)[1].lower() in ['.png', '.jpg', '.jpeg']])
    msks = sorted([p for p in glob.glob(os.path.join(mask_dir,  '*')) if os.path.splitext(p)[1].lower() in ['.png', '.jpg', '.jpeg']])

    img_map = {}
    for ip in imgs:
        b = os.path.splitext(os.path.basename(ip))[0]
        img_map[b] = ip

    mask_map = {}
    for mp in msks:
        b = os.path.splitext(os.path.basename(mp))[0]
        if b.endswith('_mask'): b = b[:-5]
        mask_map[b] = mp

    pairs = []
    for b, ip in img_map.items():
        if b in mask_map:
            pairs.append((ip, mask_map[b]))
        else:
            # not paired -> skip (we already tried auto-generation above)
            pass

    return pairs

# ================= Dataset =================
class LungDataset(Dataset):
    def __init__(self, pairs, size=CFG.img_size):
        self.pairs = pairs
        self.size = size
        if len(self.pairs) == 0:
            raise RuntimeError("‚ùå Dataset empty. No matching image-mask pairs.")
    def __len__(self): return len(self.pairs)
    def __getitem__(self, idx):
        img_path, msk_path = self.pairs[idx]
        img = Image.open(img_path).convert('L').resize((self.size, self.size))
        msk = Image.open(msk_path).convert('L').resize((self.size, self.size))
        img = np.array(img, dtype=np.float32) / 255.0
        msk = np.array(msk, dtype=np.float32) / 255.0
        img = torch.from_numpy(img).unsqueeze(0)  # [1,H,W]
        msk = torch.from_numpy(msk).unsqueeze(0)  # [1,H,W]
        return img, msk

# ================= Model: SwinUNet-ish + CBAM =================
class CBAM(nn.Module):
    def __init__(self, ch, r=16):
        super().__init__()
        self.mlp = nn.Sequential(
            nn.Linear(ch, ch//r), nn.ReLU(inplace=True),
            nn.Linear(ch//r, ch)
        )
        self.spatial = nn.Conv2d(2, 1, kernel_size=7, padding=3)
    def forward(self, x):
        b, c, h, w = x.shape
        avg = torch.mean(x, dim=(2,3), keepdim=True)
        mx  = torch.amax(x, dim=(2,3), keepdim=True)
        ch_att = torch.sigmoid(self.mlp(avg.view(b,c)) + self.mlp(mx.view(b,c))).view(b,c,1,1)
        x = x * ch_att
        avg2 = torch.mean(x, dim=1, keepdim=True)
        mx2  = torch.max(x, dim=1, keepdim=True)[0]
        sp_att = torch.sigmoid(self.spatial(torch.cat([avg2, mx2], dim=1)))
        return x * sp_att

class BasicBlock(nn.Module):
    def __init__(self, ch):
        super().__init__()
        self.block = nn.Sequential(
            nn.Conv2d(ch, ch, 3, padding=1), nn.BatchNorm2d(ch), nn.ReLU(inplace=True),
            nn.Conv2d(ch, ch, 3, padding=1), nn.BatchNorm2d(ch), nn.ReLU(inplace=True),
            nn.Dropout(CFG.dropout_p)  # <-- enables Monte Carlo Dropout
        )
        self.cbam = CBAM(ch)
    def forward(self, x):
        return self.cbam(self.block(x))

class Down(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.seq = nn.Sequential(
            nn.MaxPool2d(2),
            nn.Conv2d(in_ch, out_ch, 3, padding=1),
            nn.ReLU(inplace=True),
            BasicBlock(out_ch)
        )
    def forward(self, x): return self.seq(x)

class Up(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.up = nn.ConvTranspose2d(in_ch, in_ch//2, kernel_size=2, stride=2)
        self.conv = nn.Conv2d(in_ch, out_ch, 3, padding=1)
        self.block = BasicBlock(out_ch)
    def forward(self, x1, x2):
        x1 = self.up(x1)
        # pad if needed
        diffY = x2.size(2) - x1.size(2)
        diffX = x2.size(3) - x1.size(3)
        x1 = F.pad(x1, [diffX//2, diffX - diffX//2, diffY//2, diffY - diffY//2])
        x = torch.cat([x2, x1], dim=1)
        return self.block(self.conv(x))

class OutConv(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__(); self.conv = nn.Conv2d(in_ch, out_ch, 1)
    def forward(self, x): return self.conv(x)

class SwinUNet_CBAM(nn.Module):
    """
    A UNet-like backbone with CBAM in blocks.
    (Compact stand-in for Swin-UNet-style encoder-decoder; keeps your CBAM+Dropout idea.)
    """
    def __init__(self):
        super().__init__()
        self.stem  = nn.Sequential(nn.Conv2d(CFG.in_ch, 64, 3, padding=1), nn.ReLU(inplace=True), BasicBlock(64))
        self.down1 = Down(64, 128)
        self.down2 = Down(128, 256)
        self.down3 = Down(256, 512)
        self.up1   = Up(512, 256)
        self.up2   = Up(256, 128)
        self.up3   = Up(128, 64)
        self.head  = OutConv(64, CFG.out_ch)
    def forward(self, x):
        x1 = self.stem(x)
        x2 = self.down1(x1)
        x3 = self.down2(x2)
        x4 = self.down3(x3)
        x  = self.up1(x4, x3)
        x  = self.up2(x,  x2)
        x  = self.up3(x,  x1)
        return torch.sigmoid(self.head(x))

# ================= Metrics =================
def dice(pred, target, eps=1e-6):
    pred = (pred > 0.5).float()
    inter = (pred * target).sum()
    return (2*inter + eps) / (pred.sum() + target.sum() + eps)

def iou(pred, target, eps=1e-6):
    pred = (pred > 0.5).float()
    inter = (pred * target).sum()
    union = pred.sum() + target.sum() - inter
    return (inter + eps) / (union + eps)

# ================= Training / Eval =================
def train_one_epoch(model, loader, opt):
    model.train()
    tloss = 0.0
    for imgs, masks in tqdm(loader, desc='Train', leave=False):
        imgs, masks = imgs.to(CFG.device), masks.to(CFG.device)
        opt.zero_grad()
        out = model(imgs)
        loss = F.binary_cross_entropy(out, masks)
        loss.backward()
        opt.step()
        tloss += loss.item()
    return tloss / max(1, len(loader))

@torch.no_grad()
def evaluate(model, loader):
    model.eval()
    dices, ious = [], []
    all_preds, all_tgts = [], []
    for imgs, masks in tqdm(loader, desc='Val', leave=False):
        imgs, masks = imgs.to(CFG.device), masks.to(CFG.device)
        out = model(imgs)
        dices.append(dice(out, masks).item())
        ious.append(iou(out, masks).item())
        all_preds.extend(out.cpu().numpy().ravel())
        all_tgts.extend(masks.cpu().numpy().ravel())

    y_true = (np.array(all_tgts) > 0.5).astype(int)
    y_pred = (np.array(all_preds) > 0.5).astype(int)
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, zero_division=0)
    rec = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)
    cm = confusion_matrix(y_true, y_pred)
    return {
        'dice': float(np.mean(dices)),
        'iou': float(np.mean(ious)),
        'acc': acc, 'prec': prec, 'rec': rec, 'f1': f1, 'cm': cm
    }

@torch.no_grad()
def mc_dropout_evaluate(model, loader, T=CFG.mc_passes):
    """
    Monte Carlo Dropout: run T stochastic passes with dropout enabled at eval time.
    Returns mean metrics across T predictions.
    """
    # enable dropout layers in eval by setting them to train mode only (keep BN etc. in eval)
    def enable_dropout(m):
        if isinstance(m, nn.Dropout) or isinstance(m, nn.Dropout2d):
            m.train()
    model.eval()
    model.apply(enable_dropout)

    dices, ious = [], []
    all_probs, all_tgts = [], []

    for imgs, masks in tqdm(loader, desc=f'MC-DO Val (T={T})', leave=False):
        imgs, masks = imgs.to(CFG.device), masks.to(CFG.device)
        # collect T passes and average probs
        probs_T = []
        for _ in range(T):
            out = model(imgs)              # dropout active
            probs_T.append(out)
        probs = torch.stack(probs_T, dim=0).mean(0)  # [B,1,H,W]
        dices.append(dice(probs, masks).item())
        ious.append(iou(probs, masks).item())
        all_probs.extend(probs.cpu().numpy().ravel())
        all_tgts.extend(masks.cpu().numpy().ravel())

    y_true = (np.array(all_tgts) > 0.5).astype(int)
    y_pred = (np.array(all_probs) > 0.5).astype(int)
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, zero_division=0)
    rec = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)
    cm = confusion_matrix(y_true, y_pred)
    return {
        'dice': float(np.mean(dices)),
        'iou': float(np.mean(ious)),
        'acc': acc, 'prec': prec, 'rec': rec, 'f1': f1, 'cm': cm
    }

def plot_last_epoch_subplot(metrics_dict, title_prefix="Last Epoch"):
    """
    Single row subplot: [Confusion Matrix | Acc | Prec | Rec | F1]
    """
    cm = metrics_dict['cm']
    acc, prec, rec, f1 = metrics_dict['acc'], metrics_dict['prec'], metrics_dict['rec'], metrics_dict['f1']

    fig, axes = plt.subplots(1, 5, figsize=(22, 4))

    # Confusion Matrix
    ax = axes[0]
    im = ax.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    ax.set_title('Confusion Matrix')
    plt.colorbar(im, ax=ax, fraction=0.046, pad=0.04)
    classes = ['BG', 'Lesion']
    ax.set_xticks([0,1]); ax.set_xticklabels(classes)
    ax.set_yticks([0,1]); ax.set_yticklabels(classes)
    thresh = cm.max()/2 if cm.max() > 0 else 0.5
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, f"{cm[i,j]}", ha='center',
                    color='white' if cm[i,j] > thresh else 'black')
    ax.set_ylabel('True'); ax.set_xlabel('Pred')

    # Bars
    labels = ['Accuracy', 'Precision', 'Recall', 'F1']
    vals   = [acc, prec, rec, f1]
    for k in range(4):
        axes[k+1].bar([labels[k]], [vals[k]])
        axes[k+1].set_ylim(0,1)
        axes[k+1].set_title(labels[k])
        axes[k+1].set_ylabel('Score')

    fig.suptitle(f"{title_prefix} ‚Äî Dice {metrics_dict.get('dice',0):.3f}, IoU {metrics_dict.get('iou',0):.3f}")
    fig.tight_layout()
    plt.show()

# ================= Main =================
def main():
    _ensure_dirs()
    if IN_COLAB:
        upload_images_and_optional_masks()

    # discover pairs for train set
    pairs = discover_pairs(CFG.img_dir, CFG.msk_dir)
    random.shuffle(pairs)

    # split train/val if val folders are empty; else use explicit val folders
    val_pairs_explicit = discover_pairs(CFG.vimg_dir, CFG.vmsk_dir)
    if len(val_pairs_explicit) == 0:
        split = int(len(pairs) * (1 - CFG.val_split))
        train_pairs, val_pairs = pairs[:split], pairs[split:]
    else:
        train_pairs, val_pairs = pairs, val_pairs_explicit

    if len(train_pairs) == 0 or len(val_pairs) == 0:
        raise RuntimeError("‚ùå Dataset empty. Add images & masks to ./data/images and ./data/masks (and/or val folders).")

    print(f"Train pairs: {len(train_pairs)} | Val pairs: {len(val_pairs)}")

    train_loader = DataLoader(LungDataset(train_pairs), batch_size=CFG.batch_size, shuffle=True, num_workers=0)
    val_loader   = DataLoader(LungDataset(val_pairs),   batch_size=CFG.batch_size, shuffle=False, num_workers=0)

    model = SwinUNet_CBAM().to(CFG.device)
    opt   = torch.optim.Adam(model.parameters(), lr=CFG.lr)
    sched = torch.optim.lr_scheduler.CosineAnnealingWarmRestarts(opt, T_0=5)

    best_dice, patience = 0.0, 0
    last_epoch_metrics = None

    for ep in range(1, CFG.epochs + 1):
        tloss = train_one_epoch(model, train_loader, opt)
        metrics = evaluate(model, val_loader)
        last_epoch_metrics = metrics
        print(f"Epoch {ep:02d}/{CFG.epochs} | Train {tloss:.4f} | "
              f"Dice {metrics['dice']:.4f} | IoU {metrics['iou']:.4f} | "
              f"Acc {metrics['acc']:.4f} | Prec {metrics['prec']:.4f} | Rec {metrics['rec']:.4f} | F1 {metrics['f1']:.4f}")

        # save best
        if metrics['dice'] > best_dice:
            best_dice = metrics['dice']; patience = 0
            torch.save(model.state_dict(), os.path.join(CFG.out_dir, 'best.pth'))
        else:
            patience += 1
        if patience >= CFG.patience:
            print("Early stopping.")
            break
        sched.step(ep)

    # ---- Last epoch subplot (CM + 4 bars) ----
    plot_last_epoch_subplot(last_epoch_metrics, title_prefix="Last Epoch (Single-pass)")

    # ---- MC-Dropout evaluation (optional but recommended) ----
    mc_metrics = mc_dropout_evaluate(model, val_loader, T=CFG.mc_passes)
    print(f"[MC-Dropout] Dice {mc_metrics['dice']:.4f} | IoU {mc_metrics['iou']:.4f} | "
          f"Acc {mc_metrics['acc']:.4f} | Prec {mc_metrics['prec']:.4f} | Rec {mc_metrics['rec']:.4f} | F1 {mc_metrics['f1']:.4f}")
    plot_last_epoch_subplot(mc_metrics, title_prefix=f"MC-Dropout (T={CFG.mc_passes})")

    torch.save(model.state_dict(), os.path.join(CFG.out_dir, 'last.pth'))
    print("‚úÖ Done. Models saved to:", CFG.out_dir)

if __name__ == "__main__":
    main()
```
